```python
# Importing the dataset 
# Function for importing data into python
# Returns a Dataset object if valid extension 
def import_dataset(filename, headerfile = '', first_row_as_headers = False):
    
    
    # Checking whether a text file for names has been added, and if so, create list of names 
    header_list = []
    
    if '.txt' in headerfile.lower():
        print('Text File inputted')
        x = 'Yes'
    elif len(headerfile) < 1:
        print('No text file for names added')
        x = 'No'
    else:
        print('Please enter Text File')
        x = 'No'
        
    try:
        if 'Yes' in x: 
            with open(headerfile, 'r') as f: 
                lines = f.readlines()
                for i in lines: 
                    header_list.append(i.replace(',', '').replace('\n', '').replace(' ', ''))    
                print('Header List successfully created')
        elif 'No' in x: 
            print('No names file added')
    except:
        print('Error with text file')

    # Checking and importing CSV File              
    if '.csv' in filename: 
        print(f'{filename} is a .csv file')
        if len(headerfile) >= 1: 
            if 'Yes' in x:
                print('Header file added to column names')
                return pd.read_csv(filename, sep = None, names = header_list, engine = 'python', header = None)
            elif 'No' in x:
                if first_row_as_headers == False: 
                    print('If inputted header file, not added')
                    return pd.read_csv(filename, sep = None, engine = 'python', header = None)
                elif first_row_as_headers == True: 
                    print('First row added as headers')
                    df = pd.read_csv(filename , sep = None, engine = 'python', header = None)
                    df.columns = df.iloc[0] 
                    return df
        elif first_row_as_headers == False: 
             print('No header file added')
             return pd.read_csv(filename, sep = None, engine = 'python', header = None)
        else:
             print('No header file added - first row as column names')
             df = pd.read_csv(filename , sep = None, engine = 'python', header = None)
             df.columns = df.iloc[0] 
             return df[1:]   
        
    elif '.xlsx' in filename: 
            print(f'{filename} is a .xlsx file')
            if len(headerfile) >= 1: 
                 if 'Yes' in x:
                    print('Header file added to column names')
                    return pd.read_excel(filename, names = header_list, header = None)
                 elif 'No' in x:
                    if first_row_as_headers == False: 
                        print('If inputted header file, not added')
                        return pd.read_excel(filename, header = None)
                    elif first_row_as_headers == True: 
                        print('First row added as headers')
                        df = pd.read_excel(filename, header = None)
                        df.columns = df.iloc[0] 
                        return df
            elif first_row_as_headers == False: 
                print('No header file added')
                return pd.read_excel(filename, header = None)
            else:
                print('No header file added - first row as column names')
                df = pd.read_excel(filename , header = None)
                df.columns = df.iloc[0] 
                return df[1:]
            
            
    elif '.data' in filename: 
             print(f'{filename} is a .data file')
             if len(headerfile) >= 1: 
                if 'Yes' in x:
                    print('Header file added to column names')
                    return pd.read_csv(filename, sep = None, names = header_list, engine = 'python', header = None)
                elif 'No' in x:
                    if first_row_as_headers == False: 
                        print('If inputted header file, not added')
                        return pd.read_csv(filename, sep = None, engine = 'python', header = None)
                    elif first_row_as_headers == True: 
                        print('First row added as headers')
                        df = pd.read_csv(filename , sep = None, engine = 'python', header = None)
                        df.columns = df.iloc[0] 
                        return df
             elif first_row_as_headers == False: 
                print('No header file added')
                return pd.read_csv(filename, sep = None, engine = 'python', header = None)
             else:
                print('No header file added - first row as column names')
                df = pd.read_csv(filename , sep = None, engine = 'python', header = None)
                df.columns = df.iloc[0] 
                return df[1:]        

```
